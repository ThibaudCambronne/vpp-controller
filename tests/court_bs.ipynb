```python
# Create a plot for "INTERVALSTARTTIME_GMT" vs MW between 2025-04-10 and 2025-04-13

def plotInterval(df, startDt, endDt):
    # Filter the dataframe for the specified date range
    mask = (df['INTERVALSTARTTIME_GMT'] >= startDt) & (df['INTERVALSTARTTIME_GMT'] <= endDt)
    filtered_df = df.loc[mask]

    # Plotting
    plt.figure(figsize=(12, 6))
    plt.plot(filtered_df['INTERVALSTARTTIME_GMT'], filtered_df['MW'], marker='o')
    plt.title('MW vs INTERVALSTARTTIME_GMT')
    plt.xlabel('INTERVALSTARTTIME_GMT')
    plt.ylabel('MW')
    plt.xticks(rotation=45)
    plt.grid()
    plt.tight_layout()
    plt.show()
    
```

```python
from datetime import date


def import_clean_save(orig_filename, new_filename):
    df = pd.read_csv(f"../data/unclean_price_data/{orig_filename}.csv")
    # Only keep the columns OPR_DT, OPR_HR, and MW
    df = df[["OPR_DT", "OPR_HR", "MW"]]

    # rename the columns from "MW" to "$/MW"
    df = df.rename(columns={"MW": "$/MW"})
    
    # Save the filtered dataframe to a new CSV file in data folder with the specified filename
    filepath = f"../data/{new_filename}"
    df.to_csv(filepath, index=False)

import_clean_save("bad_0096WD_7_N001_fall_2025_10_10", "pricedf_0096WD_7_N001_fall_2025_10_10.csv")
import_clean_save("bad_0096WD_7_N001_spring_2025_04_10", "pricedf_0096WD_7_N001_spring_2025_04_10.csv")
import_clean_save("bad_0096WD_7_N001_summer_2025_07_10", "pricedf_0096WD_7_N001_summer_2025_07_10.csv")
import_clean_save("bad_0096WD_7_N001_winter_2025_01_10", "pricedf_0096WD_7_N001_winter_2025_01_10.csv")
```
