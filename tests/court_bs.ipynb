### Create Price Dataframes for specific dates

```python
price_df = pd.read_csv(f"../data/0096WD_7_N001_LMP_12months.csv")

# Sort dataframe by date times INTERVALSTARTTIME_GMT
price_df = price_df.sort_values(by="INTERVALSTARTTIME_GMT")


price_df
```

```python
def filter_by_day_and_save(df, date, filename):
    # Only keep the columns INTERVALSTARTTIME_GMT and MW
    df = df[["INTERVALSTARTTIME_GMT", "MW"]]

    # rename the columns from "INTERVALSTARTTIME_GMT" to "StartTime_GMT" and "MW" to "$/MW"
    df = df.rename(columns={"INTERVALSTARTTIME_GMT": "StartTime_GMT", "MW": "$/MW"})

    

    # Convert INTERVALSTARTTIME_GMT to datetime
    df["StartTime_GMT"] = pd.to_datetime(df["StartTime_GMT"])

    # Filter the dataframe for the specified date
    filtered_df = df[df["StartTime_GMT"].dt.date == pd.to_datetime(date).date()].reset_index(drop=True)
    
    display(filtered_df)
    
    # Save the filtered dataframe to a new CSV file in data folder with the specified filename
    filepath = f"../data/{filename}"
    filtered_df.to_csv(filepath, index=False)

filter_by_day_and_save(price_df, "2025-04-10", "price_df_0096WD_7_N001_2025-04-10.csv")
filter_by_day_and_save(price_df, "2025-04-10", "price_df_0096WD_7_N001_2025-04-10.csv")
filter_by_day_and_save(price_df, "2025-04-10", "price_df_0096WD_7_N001_2025-04-10.csv")
filter_by_day_and_save(price_df, "2025-04-10", "price_df_0096WD_7_N001_2025-04-10.csv")
```
